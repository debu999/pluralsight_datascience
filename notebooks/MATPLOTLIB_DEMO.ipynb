```python
import  matplotlib.pyplot as plt
%matplotlib inline
# import os
# print(os.path.join(os.getcwd(), os.pardir,"src","data"))
# os.listdir(os.path.join(os.getcwd(), os.pardir,"src","data"))
```

```python
import sys,os;
import get_processed_data
sys.path.append(os.path.join(os.getcwd(), os.pardir,"src","data"));
from get_processed_data import read_data as rd, process_data as  pd
df = rd()
df = pd(df)
df.info()
```

```python
plt.hist(df.Age, bins=15, color="darkgreen");
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution Chart")
plt.show()
```

```python
fig, axs = plt.subplots()
axs.hist(df.Age, bins=15, color="darkgreen");
axs.set_xlabel("Age")
axs.set_ylabel("Count")
axs.set_title("Age Distribution Chart")
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].hist(df.Age, bins=15, color="darkgreen",edgecolor="white",linewidth=1.3);
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Count")
axs[0].set_title("Age Distribution Chart")
axs[1].hist(df.Fare, bins=15, color="tomato",edgecolor="white",linewidth=1.3);
axs[1].set_xlabel("Age")
axs[1].set_ylabel("Count")
axs[1].set_title("Fare Distribution Chart")
plt.show()
```

```python
fig, axs = plt.subplots(3,2,figsize=(25,25))
axs[0,0].hist(df.Age, bins=15, color="darkgreen",edgecolor="white",linewidth=1.3);
axs[0,0].set_xlabel("Age")
axs[0,0].set_ylabel("Count")
axs[0,0].set_title("Age Distribution Chart")
axs[0,1].hist(df.Fare, bins=15, color="maroon",edgecolor="white",linewidth=1.3);
axs[0,1].set_xlabel("Age")
axs[0,1].set_ylabel("Count")
axs[0,1].set_title("Fare Distribution Chart")
axs[1,0].boxplot(df.Fare.values);
axs[1,0].set_xlabel("Fare")
axs[1,0].set_ylabel("Fare")
axs[1,0].set_title("Box Plot: Fare")
axs[1,1].boxplot(df.Age.values);
axs[1,1].set_xlabel("Age")
axs[1,1].set_ylabel("Age")
axs[1,1].set_title("Box Plot: Age")
axs[2,0].scatter(x=df.Age,y=df.Fare,color="darkblue", alpha=0.5, linewidths = 1, edgecolors ="white");
axs[2,0].set_xlabel("Age")
axs[2,0].set_ylabel("Fare")
axs[2,0].set_title("Age vs Fare")
plt.tight_layout()
axs[2,1].axis("off")
plt.show()
```
