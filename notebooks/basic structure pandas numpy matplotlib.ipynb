```python
def df_desc(df,col_list=[0,1,3,5,6,8,9,]):
    desc_pd=pd.DataFrame({"count":df.iloc[:,col_list].count(),
                          "mean":df.iloc[:,col_list].mean(),
                          "std":df.iloc[:,col_list].std(),
                          "min":df.iloc[:,col_list].min(),
                          "max":df.iloc[:,col_list].max(),
                          "var":df.iloc[:,col_list].var(),
                          "range":df.iloc[:,col_list].max()-df.iloc[:,col_list].min(),
                          "25%":df.iloc[:,col_list].quantile(.25),
                          "50%":df.iloc[:,col_list].quantile(.5),
                          "75%":df.iloc[:,col_list].quantile(.75)}).T
    return desc_pd.loc[["count","mean","std","min","var","range","25%","50%","75%","max"],:]
df_desc(titanic_df)
```
